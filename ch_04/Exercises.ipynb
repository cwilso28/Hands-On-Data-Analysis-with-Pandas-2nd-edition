## Exercise 9

9. Use the `transform()` method on the FAANG data to represent all the values in terms of the first date in the data. To do so, divide all the values for each ticker by the values for the first date in the data for that ticker. This is referred to as an index, and the data for the first date is the base (https://ec.europa.eu/eurostat/statistics-explained/index.php/Beginners:Statistical_concept__Index_and_base_year). When data is in this format, we can easily see growth over time. Hint: `transform()` can take a function name.

```python
faang.groupby('ticker')[['high', 'low', 'open', 'close', 'volume']].transform(lambda x: x/x[x.index.min()])
```

## Part A

```python
covid19 = pd.read_csv('./exercises/covid19_cases.csv', parse_dates = ['dateRep'], index_col = 'dateRep').replace({'United_States_of_America': 'USA', 'United_Kingdom':'UK'}).sort_index()
```

## Part B

```python
top5countries = covid19.groupby('countriesAndTerritories').agg({'cases': 'sum'}).nlargest(5, 'cases').index.values
top5covid = covid19[covid19.countriesAndTerritories.str.contains('|'.join(index for index in top5countries))]
```

```python
top5covid.nlargest(1, 'cases')
```

## Part C

```python
top5covid.groupby('countriesAndTerritories').last('1W')
```

```python
top5covid['rolling_cases'] = top5covid.groupby('countriesAndTerritories')['cases'].transform(lambda x: x.rolling('7D').mean())
```

```python
top5covid.last('1W')
```

## Part D

```python
covid19[(covid19.countriesAndTerritories != 'China') & (covid19.cases > 0)].reset_index().groupby('countriesAndTerritories').first().reset_index().set_index('dateRep')
```
