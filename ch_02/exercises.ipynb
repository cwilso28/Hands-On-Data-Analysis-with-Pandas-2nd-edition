```python
import pandas as pd
import numpy as np

df = pd.read_csv('./data/earthquakes.csv')
df.head()
```

1. Find the 95th percentile of earthquake magnitude in Japan using the `mb` magnitude type.

```python
df.columns
```

```python
df_japan_mb = df.loc[(df.magType == 'mb') & (df.place.str.contains('Japan')), ['place', 'mag', 'magType']]
df_japan_mb.describe(percentiles =[0.95])
df_japan_mb.mag.quantile(0.95)
```

```python
df[(df.magType == 'mb') & (df.place.str.contains('Japan'))].mag.quantile(0.95)
```

2. Find the percentage of earthquakes in Indonesia that were coupled with tsunamis.

```python
df_ind = df[df.place.str.contains('Indonesia')]
tsunami_count = sum(df_ind.tsunami)
print('{0:0.2f}%'.format(100*tsunami_count / df_ind.shape[0]))
```

```python
f"{df[df.place.str.contains('Indonesia')].tsunami.value_counts(normalize=True).iloc[1,]:.2%}"
```
